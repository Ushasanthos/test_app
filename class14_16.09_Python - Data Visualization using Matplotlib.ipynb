```python
import numpy as np
import pandas as pd
# from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
# Graph paper
# Draw axes (x axis and y axis)
# Plotting
#   Put points/coordinates
#   Connect the points
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
fig = plt.figure()                     # fig_size = 640*480
ax = fig.add_axes([0, 0, 0.5, 0.5])    # Origin x, Origin y, Ratio x axis, Ratio y axis w.r.t. figure
ax.plot(x, y)
ax.set_title('Sine Function')
ax.set_xlabel('Angle')
ax.set_ylabel('Sine of Angle')
```

```python
x = np.arange(0, 2*np.pi, 0.1)
z = np.cos(x)
fig = plt.figure()                     
ax = fig.add_axes([0, 0, 0.5, 0.5])    
ax.plot(x, z)
ax.set_title('Cosine Function')
ax.set_xlabel('Angle')
ax.set_ylabel('Cosine of Angle')
```

```python
x = np.arange(0, 2*np.pi, 0.1)
l = np.log10(x)
fig = plt.figure()                     
ax = fig.add_axes([0, 0, 0.5, 0.5])    
ax.plot(x, l)
ax.set_title('Logrithmic Function')
ax.set_xlabel('Data')
ax.set_ylabel('Log of Data')
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()                     
ax = fig.add_axes([0, 0, 0.5, 0.5])    

ax.plot(x, y, label = 'Sine')
ax.plot(x, z, label = 'Cosine')
ax.set_title('Trigonometric Function')
ax.legend()
```

### Mul#tiplotting

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()                     
ax = fig.add_axes([0, 0, 0.5, 0.5])    

ax.plot(x, y, label = 'Sine', color = 'blue')
ax.plot(x, z, label = 'Cosine', color = '#d0312d')
ax.set_title('Trigonometric Function')
ax.legend()
```

#### Subplotting

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(1, 2) 
plt.tight_layout()

ax[0].plot(x, y, label = 'Sine', color = 'blue')
ax[0].plot(x, z, label = 'Cosine', color = '#d0312d')
ax[0].set_title('Trigonometric Function')
ax[0].legend()

ax[1].plot(x, np.power(x, 2), label = 'Square', color = 'blue')
ax[1].plot(x, np.power(x, 3), label = 'Cube', color = 'red', marker = '*')
ax[1].set_title('Power Function')
ax[1].legend()
```

```python
states = ['HP', 'UP', 'MP', 'AP']
popul =  [3,    25,    18,   16]
print("Population in these 4 states: ", sum(popul))

def absolute_value(val):
    return np.round(val/100.*np.sum(popul))

plt.pie(popul, labels = states, autopct = absolute_value)
plt.show()
```
