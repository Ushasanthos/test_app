```python
x = np.arange(0, np.pi, 0.1)
y = np.sin(x)
fig = plt.figure()                     # fig_size = 640*480
ax = fig.add_axes([0, 0, 0.8, 0.8])    # Origin x, Origin y, Ratio x axis, Ratio y axis w.r.t. figure
ax.plot(x, y)
ax.set_title('Sine Function')
ax.set_xlabel('Angle')
ax.set_ylabel('Sine of Angle')
```

```python
x = np.arange(0, 2*np.pi, 0.1)
z = np.cos(x)
fig = plt.figure()                     
ax = fig.add_axes([0, 0, 0.5, 0.5])    
ax.plot(x, z)
ax.set_title('Cosine Function')
ax.set_xlabel('Angle')
ax.set_ylabel('Cosine of Angle')
```

```python
x = np.arange(0, 2*np.pi, 0.1)
l = np.log10(x)
fig = plt.figure()                     
ax = fig.add_axes([0, 0, 0.5, 0.5])    
ax.plot(x, l)
ax.set_title('Logrithmic Function')
ax.set_xlabel('Data')
ax.set_ylabel('Log of Data')
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()                     
ax = fig.add_axes([0, 0, 0.5, 0.5])    

ax.plot(x, y, label = 'Sine')
ax.plot(x, z, label = 'Cosine')
ax.set_title('Trigonometric Function')
ax.legend()
```

#### Subplotting

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(1, 2) 
plt.tight_layout()

ax[0].plot(x, y, label = 'Sine', color = 'blue')
ax[0].plot(x, z, label = 'Cosine', color = '#d0312d')
ax[0].set_title('Trigonometric Function')
ax[0].legend()

ax[1].plot(x, np.power(x, 2), label = 'Square', color = 'blue')
ax[1].plot(x, np.power(x, 3), label = 'Cube', color = 'red', marker = '*')
ax[1].set_title('Power Function')
ax[1].legend()
```

```python
def add(x, y):
    x = int(input('First: '))
    y = int(input('Second: '))
    return x + y

def subtract(x, y):
    x = int(input('First: '))
    y = int(input('Second: '))
    return x - y

def multiply(x, y):
    x = int(input('First: '))
    y = int(input('Second: '))
    return x * y

op = input('Enter operator (+,-,*): ')
if   op == '+': print(add())
elif op == '-': print(subtract())
elif op == '*': print(multiply())
else          : print('Invalid input')
    
```

```python
def add():
    x = int(input('First: '))
    y = int(input('Second: '))
    return x + y

def subtract():
    x = int(input('First: '))
    y = int(input('Second: '))
    return x - y

def multiply():
    x = int(input('First: '))
    y = int(input('Second: '))
    return x * y

d = {'+': add, '-': subtract, '*': multiply}
op = input('Enter operator (+,-,*): ')
print(d[op]())            # d[op] is add | d[op]() is add()
```

```python
def ageCat(a):
    ne = np.linspace(n.Age.min(), n.Age.max(), 6)
    if   ne[0] <= a < ne[1]: return 'A'
    elif ne[1] <= a < ne[2]: return 'B'
    elif ne[2] <= a < ne[3]: return 'C'
    elif ne[3] <= a < ne[4]: return 'D'
    elif ne[4] <= a <=ne[5]: return 'E'
    
n['AgeGroup'] = n.Age.apply(ageCat)
new = n.AgeGroup.value_counts()
new.sort_index(inplace = True)

fig = plt.figure()                     
ax = fig.add_axes([0, 0, 0.5, 0.5]) 

ax.bar(new.index, new, width = 1.0)
```

#### Analysis - based on number of variables
- Univariate Analysis: Single Variable
    - Understanding trend of data of individual columns. Descriptive Statistics
- Bivariate Analysis: Two Variable: 
    - Understanding their relationship/interdependency
- Multivariate Analysis: More than two variable. 
    - Understanding entire dataset and features interdependency

### Correlation Matrix
A correlation matrix is a square matrix showing the correlation coefficients between two variables. Correlation coefficients measure how strong and in which direction two variables are linked in a straight line. A correlation matrix often examines how different variables relate in multivariate analysis and statistics.

### Jointplot
- This plot helps you to analyze two columns of continuous data.
- It will provide you the insights or patterns between them.
- It shows the comparison between two columns along with their distribution plot.
- it is a combination of two plots. Inner plot is always scatter. The outer one is default to displot
