### Merging Data in Pandas

```python
l = pd.DataFrame({'id':[1,2,3,4,5], 
                  'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
                  'subject_id':['sub1','sub2','sub4','sub6','sub5']})

r = pd.DataFrame({'id':[1,2,3,4,5], 
                  'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
                  'subject_id':['sub2','sub4','sub3','sub6','sub5']})
```

```python
l
```

```python
r
```

```python
# id        : mutually inclusive
# name      : mutually exclusive
# subject_id: partially inclusive
```

```python
t = pd.merge(l, r)
t
```

```python
pd.merge(l, r, on = 'id')
```

```python
pd.merge(l, r, on = 'Name')      # how = 'inner' which means intersection
```

```python
pd.merge(l, r, on = 'subject_id')   # how = 'inner'
```

```python
l
```

```python
r
```

```python
pd.merge(l, r, on = 'subject_id', how = 'outer')    # outer means union
```

```python
pd.merge(l, r, on = 'subject_id', how = 'left')
```

```python
pd.merge(l, r, on = 'subject_id', how = 'right')
```

```python
pd.merge(l, r, on = ['id','subject_id'], how = 'inner')
```

```python
pd.merge(l, r, on = ['subject_id'], how = 'inner')
```
